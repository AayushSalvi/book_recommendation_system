# Content Based Recommeding System 

```python
new_df = data.copy()
new_df['tags'] = new_df['Author']+new_df['Description']+new_df['Genres']+new_df['Avg_Rating'].astype(str)
```

```python
new_df = new_df[['Book','tags']]
```

```python
#new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
new_df['tags'] =new_df['tags'].str.replace('\W', ' ', regex=True)
```

```python
new_df['tags']
```

```python
new_df.shape
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
arr = cv.get_feature_names_out()
```

```python
arr[0:1000]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
new_df
```

```python
sorted(similarity[9922],reverse=True)[0:10]
#but the index position is lost
sorted(list(enumerate(similarity[9922])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(book):
    book_index = new_df[new_df['Book']==book].index[0]
    distances = similarity[book_index]
    book_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in book_list:
        print(new_df.iloc[i[0]].Book)
        
     
```

```python
recommend('Dynamo')
```
